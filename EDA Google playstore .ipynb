**Exploratory Data Analysis (EDA)**

A. Univariate Analysis







***1..Distribution of App Categories:***



```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Apps in Each Category')
plt.ylabel('')
plt.show()
```


**2..Distribution of App Ratings:  **

```python
.plt.figure(figsize=(10,5))
df['Rating'].plot(kind='hist', bins=20)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

**B. Bivariate Analysis**


**1..Relationship Between Rating and Number of Reviews**:



```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Reviews', y='Rating', data=df)
plt.title('Reviews vs Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.show()
```

**2.Price vs. Installs:**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Price', y='Installs', data=df)
plt.title('Price vs Installs')
plt.xlabel('Price ($)')
plt.ylabel('Number of Installs')
plt.show()
```

**3.Box Plot of Ratings by Category:**


```python
plt.figure(figsize=(14,8))
sns.boxplot(x='Category', y='Rating', data=df)
plt.xticks(rotation=90)
plt.title('Box Plot of Ratings by Category')
plt.show()
```

**C. Correlation Analysis**

**1.Correlation Matrix:**


```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
```
